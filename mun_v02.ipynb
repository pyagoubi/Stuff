```python
import pandas as pd




file_names = ['bakery']

# , 'bar', 'education', 'grocery', 'kiosk', 'post',
#               'restaurant', 'veloparking', 'pharmacy', 'public_park',
#               'coworking', 'haltestelle']



# Load gmd dataframe (replace 'gmd.csv' with the actual file or path)
gmd_df = pd.read_csv('gmd.csv')

file_names = ['bakery', 'bar', 'education', 'grocery', 'kiosk', 'post',
              'restaurant', 'veloparking', 'pharmacy', 'public_park',
              'coworking', 'haltestelle']

def filter_dataframes(muns_df, gmd_df, file_names):
    # Dictionary to hold all the clean and dirty dataframes
    dataframes = {}

    total_files = len(file_names)  # Total number of files
    current_file = 0  # Current file number

    for file_name in file_names:
        current_file += 1  # Increment the file number
        print(f'Processing file {current_file}/{total_files}: {file_name}...')  # Status message

        # Read the current csv file into a dataframe
        df = pd.read_csv(f'{file_name}.csv')

        # Convert 'address' to string type and drop rows where 'address' does not contain 'Switzerland', case insensitive
        df = df[df['address'].astype(str).str.contains('Switzerland', case=False)]

        # Initialize clean and dirty dataframes
        clean_df = pd.DataFrame()
        dirty_df = pd.DataFrame()

        for i, row in df.iterrows():
            # Check if the municipality is in the address
            if any(municipality in row['address'] for municipality in muns_df['municipality']):
                clean_df = pd.concat([clean_df, pd.DataFrame(row).T])
            else:
                dirty_df = pd.concat([dirty_df, pd.DataFrame(row).T])

        # Go through dirty_df and check for PLZ in address
        for i, row in dirty_df.iterrows():
            for j, gmd_row in gmd_df.iterrows():
                if gmd_row['Ortschaftsname'] in muns_df['municipality'].tolist():
                    if str(gmd_row['PLZ']) in row['address']:
                        # If PLZ is in address, move row from dirty to clean df
                        clean_df = pd.concat([clean_df, pd.DataFrame(row).T])
                        dirty_df = dirty_df.drop(i)

        # Add the clean and dirty dataframes to the dictionary
        dataframes[f'{file_name}_clean'] = clean_df
        dataframes[f'{file_name}_dirty'] = dirty_df

        print(f'Finished processing file {current_file}/{total_files}: {file_name}.')  # Status message

    return dataframes

# Run the function
dataframes = filter_dataframes(muns_df, file_names)


# Extract each dataframe into a separate variable
bakery_clean = dataframes['bakery_clean']
bakery_dirty = dataframes['bakery_dirty']
bar_clean = dataframes['bar_clean']
bar_dirty = dataframes['bar_dirty']
education_clean = dataframes['education_clean']
education_dirty = dataframes['education_dirty']
grocery_clean = dataframes['grocery_clean']
grocery_dirty = dataframes['grocery_dirty']
kiosk_clean = dataframes['kiosk_clean']
kiosk_dirty = dataframes['kiosk_dirty']
post_clean = dataframes['post_clean']
post_dirty = dataframes['post_dirty']
restaurant_clean = dataframes['restaurant_clean']
restaurant_dirty = dataframes['restaurant_dirty']
veloparking_clean = dataframes['veloparking_clean']
veloparking_dirty = dataframes['veloparking_dirty']
pharmacy_clean = dataframes['pharmacy_clean']
pharmacy_dirty = dataframes['pharmacy_dirty']
public_park_clean = dataframes['public_park_clean']
public_park_dirty = dataframes['public_park_dirty']
coworking_clean = dataframes['coworking_clean']
coworking_dirty = dataframes['coworking_dirty']
haltestelle_clean = dataframes['haltestelle_clean']
haltestelle_dirty = dataframes['haltestelle_dirty']
```
